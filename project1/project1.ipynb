```python
# Importing required libraries for the project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plots
%matplotlib inline

print("Libraries imported successfully.")
```

```python
# Load the Regents dataset

url = "https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv"

df = pd.read_csv(url)

# Show the first five rows of the dataset
df.head()
```

```python
# STEP 1: Filter dataset to chosen school

school_name = "P.S. 034 Franklin D. Roosevelt"

school_df = df[df["School Name"] == school_name]

school_df.head()
```

```python
# STEP 2: Exams this school has

school_df["Regents Exam"].unique()
```

```python
# STEP 3: Filter to the chosen exam
exam_name = "Common Core Algebra"

school_exam_df = school_df[school_df["Regents Exam"] == exam_name]

school_exam_df
```

```python
# STEP 4: Select a metric to analyze
metric = "Total Tested"

school_exam_df[metric].describe()
```

```python
# STEP 5: Cleaning task - remove rows with suppressed or missing values
cleaned_df = school_exam_df.replace("s", pd.NA).replace("na", pd.NA)
cleaned_df = cleaned_df.dropna()

cleaned_df
```

```python
# STEP 6: Visualization - Bar chart of Total Tested across years
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(school_exam_df["Year"], school_exam_df["Total Tested"])
plt.title(f"Total Tested over Years - {school_name} ({exam_name})")
plt.xlabel("Year")
plt.ylabel("Total Tested")
plt.show()
```

```python
# STEP 7: Comparison with all NYC schools
# Compare the chosen school's average to the overall average

chosen_mean = school_exam_df["Total Tested"].mean()
overall_mean = df[df["Regents Exam"] == exam_name]["Total Tested"].replace("s", pd.NA).dropna().astype(float).mean()

print("Average Total Tested at", school_name, "for", exam_name, ":", chosen_mean)
print("Average Total Tested across ALL NYC schools for", exam_name, ":", overall_mean)
```

Based on the cleaned dataset, P.S. 034 Franklin D. Roosevelt had very small testing numbers for the Common Core Algebra exam, averaging just over two students per year. When compared with the citywide average of more than fifty students tested per school, its clear that this schools results are limited simply because of low participation. With such a small sample size, the schools Regents performance cant be meaningfully evaluated without additional years of complete data.
